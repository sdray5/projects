```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
card_df = pd.read_csv('Summoner_Wars_Cards.csv', index_col=0)
card_df
```

```python
card_df['Cost'].mean()
```

```python
events = card_df[card_df['Copies'] == 2]
events
```

```python
summoners = card_df[card_df['Card Classes'].apply(lambda x: 'Summoner' in x)]
summoners
```

```python
commons = card_df[card_df['Card Classes'].apply(lambda x: 'Common' in x)]
commons
```

```python
champions = card_df[card_df['Card Classes'].apply(lambda x: 'Champion' in x)]
champions
```

```python
w=.5
```

```python
data = events['Cost']
plt.hist(data, bins=np.arange(min(data)-.25, max(data) + w + .25, w))
# only 10 out of 80 events cost 1 magic
```

```python
data = summoners['Strength']
plt.hist(data, bins=np.arange(min(data)-.25, max(data) + w + .25, w))
# most summoners have 3 AV
```

```python
data = summoners['Life']
plt.hist(data, bins=np.arange(min(data)-.25, max(data) + w + .25, w))
# most summoners fall between 10 and 13 life
```

```python
data = commons['Cost']
plt.hist(data, bins=np.arange(min(data)-.25, max(data) + w + .25, w))
# 0-2 cost is standard
```

```python
data = commons['Strength']
plt.hist(data, bins=np.arange(min(data)-.25, max(data) + w + .25, w))
# 2 or 3 strength is standard
```

```python
data = commons['Life']
plt.hist(data, bins=np.arange(min(data)-.25, max(data) + w + .25, w))
# 2-5 life is where most commons fall
```

```python
data = champions['Cost']
plt.hist(data, bins=np.arange(min(data)-.25, max(data) + w + .25, w))
# 5 & 6 cost are pretty typical
```

```python
data = champions['Strength']
plt.hist(data, bins=np.arange(min(data)-.25, max(data) + w + .25, w))
# 2-4 is standard for champions with 3 strength being most common
```

```python
data = champions['Life']
plt.hist(data, bins=np.arange(min(data)-.25, max(data) + w + .25, w))
# 6 - 8 life pretty common. Not many champions fall below 6 life
```
