```python
 researchers_cleaned = researchers_no_missing[[not math.isnan(i) for i in pd.to_numeric(researchers_no_missing['value'], errors='coerce')]]
```

```python
np.corrcoef(pd.to_numeric(researchers_cleaned['value'], errors='coerce'), researchers_cleaned['time'])
# not a very strong correlation between time and value. suppose that makes some sense since there are many other
# variables that have larger impact
```

```python
researchers_cleaned
```

```python
yrs = researchers_cleaned.time.value_counts().index
```

```python
researchers_cleaned.time.value_counts().values
```

```python
means = []
years = []
for i in yrs.sort_values():
    means.append(pd.to_numeric(researchers_cleaned[researchers_cleaned['time']==i]['value'],errors = 'coerce').mean())
    years.append(i)
print(means)
print(years)
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
x = years
x_pos = [i for i, _ in enumerate(x)]

plt.bar(x_pos, means, color='green')
plt.xlabel("Year")
plt.ylabel("Researchers")
plt.title("Mean Researchers by year")

plt.xticks(x_pos, x)

plt.show()
```

```python
locations = researchers_cleaned.location.value_counts().index
```

```python
researchers_cleaned_2016 = researchers_cleaned[researchers_cleaned['time']==2012]
```

```python
sums_2016 = []
countries_2016 = []
for i in locations.sort_values():
    sums_2016.append(pd.to_numeric(researchers_cleaned_2016[researchers_cleaned_2016['location']==i]['value'],errors = 'coerce').sum())
    countries_2016.append(i)
```

```python
df = pd.DataFrame({'countries_2016' : countries_2016,
                   'sums_2016' : sums_2016})
```

```python
df.sort_values('sums_2016', inplace=True, ascending=False)
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
x = df['countries_2016'][:10]
x_pos = [i for i, _ in enumerate(x)]

plt.bar(x_pos, df['sums_2016'][:10], color='green')
plt.xlabel("Countries")
plt.ylabel("Sums")
plt.title("Researchers by country 2016")

plt.xticks(x_pos, x)

plt.show()
# The 10 countries with the most researchers (in this dataset)
```
