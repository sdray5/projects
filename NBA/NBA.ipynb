```python
teams = pd.read_csv('teams.csv')
teams
```

```python
team_names = pd.DataFrame([teams['TEAM_ID'], teams['CITY']+' ' + teams['NICKNAME']]).transpose()
team_names.columns = ['TEAM_ID', 'TEAM_NAME']
team_names
```

```python
games['HOME_TEAM_ID']
```

```python
merged_dataframe = pd.merge(games, team_names,
                       how='left', left_on='HOME_TEAM_ID', right_on='TEAM_ID')
merged_dataframe = pd.merge(merged_dataframe, team_names,
                       how='left', left_on='VISITOR_TEAM_ID', right_on='TEAM_ID')
merged_dataframe = merged_dataframe.rename(columns={'TEAM_NAME_y': 'VISITOR_TEAM_NAME', 'TEAM_NAME_x': 'HOME_TEAM_NAME'})
merged_dataframe = merged_dataframe.drop(columns=['TEAM_ID_x', 'TEAM_ID_y'], axis=1)
merged_dataframe
```

```python
merged_dataframe.columns
```

```python
recent_win_rate_dict = {}
for team in merged_dataframe['HOME_TEAM_NAME'].unique():
    # start them off assuming average win rate of .5
    recent_win_rate_dict[team] = .5
    
update_rate = .1
recent_home_team_win_rate = []
recent_visitor_team_win_rate = []
for index, row in merged_dataframe[::-1].iterrows():
    recent_home_team_win_rate.append(recent_win_rate_dict[row['HOME_TEAM_NAME']])
    recent_visitor_team_win_rate.append(recent_win_rate_dict[row['VISITOR_TEAM_NAME']])
    if row['HOME_TEAM_WINS'] == 1:
        winning_team = row['HOME_TEAM_NAME']
        losing_team = row['VISITOR_TEAM_NAME']    
    else:
        winning_team = row['VISITOR_TEAM_NAME']
        losing_team = row['HOME_TEAM_NAME']
    recent_win_rate_dict[winning_team] = recent_win_rate_dict[winning_team]*(1-update_rate) + update_rate
    recent_win_rate_dict[losing_team] = recent_win_rate_dict[losing_team]*(1-update_rate)    
    print(winning_team, recent_win_rate_dict[winning_team])
    print(losing_team, recent_win_rate_dict[losing_team])
```

```python
recent_home_team_win_rate = list(reversed(recent_home_team_win_rate))
recent_visitor_team_win_rate = list(reversed(recent_visitor_team_win_rate))
```

```python
merged_dataframe['RECENT_HOME_TEAM_WIN_RATE'] = recent_home_team_win_rate
merged_dataframe['RECENT_VISITOR_TEAM_WIN_RATE'] = recent_visitor_team_win_rate
merged_dataframe
```

```python
X = merged_dataframe[['RECENT_HOME_TEAM_WIN_RATE', 'RECENT_VISITOR_TEAM_WIN_RATE']]
y = merged_dataframe['HOME_TEAM_WINS']
X_test, X_train = X[:int(len(merged_dataframe)*.1)], X[int(len(merged_dataframe)*.1):]
y_test, y_train = y[:int(len(merged_dataframe)*.1)], y[int(len(merged_dataframe)*.1):]
X_train
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier()
clf.fit(X_train, y_train)
```

```python
print (clf.score(X_train, y_train))
print(clf.score(X_test, y_test))
# overfitting. let's introduce a max depth
```

```python
clf=RandomForestClassifier(max_depth=10)
clf.fit(X_train, y_train)
```

```python
print (clf.score(X_train, y_train))
print(clf.score(X_test, y_test))
# much better
```

```python
clf=RandomForestClassifier(max_depth=10, min_samples_leaf=5)
clf.fit(X_train, y_train)
```

```python
print (clf.score(X_train, y_train))
print(clf.score(X_test, y_test))
```

```python
max_depths = [None, 20, 10, 5]
min_samples_leaf_list = [1, 5, 10]
best_score = 0
for max_depth in max_depths:
    for min_samples_leaf in min_samples_leaf_list:
        clf=RandomForestClassifier(max_depth=max_depth, min_samples_leaf=min_samples_leaf)
        clf.fit(X_train, y_train)
        print(f"Scores for tree with max_depth of {max_depth} and min_samples_leaf of {min_samples_leaf}")
        print (clf.score(X_train, y_train))
        print(clf.score(X_test, y_test))
        if clf.score(X_test, y_test) > best_score:
            best_params = [max_depth, min_samples_leaf]
            best_score = clf.score(X_test, y_test)
print(f"The best parameters found were a max_depth of {best_params[0]} and min_samples_leaf of {best_params[1]}")
print(best_score)
# apparently the best set of parameters the model found had a max_depth of 5 and min_samples_leaf of 5
```

```python
from sklearn.linear_model import LogisticRegression
  
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
print (classifier.score(X_train, y_train))
print(classifier.score(X_test, y_test))
```

```python
# let's try giving the model several different update rates at once and it can decide for itself which are useful
# this dictionary will keep track of the most up to date weighted moving average win rates
recent_win_rate_dict = {}
update_rates = [.003, .01, .03, .1, .3]
for team in merged_dataframe['HOME_TEAM_NAME'].unique():
    for update_rate in update_rates:
        # start them off assuming average win rate of .5, keeping track of each team paired with update rates
        recent_win_rate_dict[team + str(update_rate)] = .5

# this dictionary will have lists of the win rates for each row in the dataframe
recent_home_team_win_rates = {}
for update_rate in update_rates:
    recent_home_team_win_rates[str(update_rate)] = []
recent_visitor_team_win_rates = {}
for update_rate in update_rates:
    recent_visitor_team_win_rates[str(update_rate)] = []
    
for index, row in merged_dataframe[::-1].iterrows():
    
    if row['HOME_TEAM_WINS'] == 1:
        winning_team = row['HOME_TEAM_NAME']
        losing_team = row['VISITOR_TEAM_NAME']    
    else:
        winning_team = row['VISITOR_TEAM_NAME']
        losing_team = row['HOME_TEAM_NAME']
    for update_rate in update_rates:
        recent_home_team_win_rates[str(update_rate)].append(recent_win_rate_dict[row['HOME_TEAM_NAME'] + str(update_rate)])
        recent_visitor_team_win_rates[str(update_rate)].append(recent_win_rate_dict[row['VISITOR_TEAM_NAME'] + str(update_rate)])
        recent_win_rate_dict[winning_team + str(update_rate)] = recent_win_rate_dict[winning_team + str(update_rate)]*(1-update_rate) + update_rate
        recent_win_rate_dict[losing_team + str(update_rate)] = recent_win_rate_dict[losing_team + str(update_rate)]*(1-update_rate) 
recent_home_team_win_rates
```

```python
for update_rate in update_rates:
    merged_dataframe[f'RECENT_HOME_TEAM_WIN_RATE_{update_rate}'] = list(reversed(recent_home_team_win_rates[str(update_rate)]))
    merged_dataframe[f'RECENT_VISITOR_TEAM_WIN_RATE_{update_rate}'] = list(reversed(recent_visitor_team_win_rates[str(update_rate)]))
merged_dataframe
```

```python
X_cols = []
for update_rate in update_rates:
    X_cols.append(f'RECENT_HOME_TEAM_WIN_RATE_{str(update_rate)}')
    X_cols.append(f'RECENT_VISITOR_TEAM_WIN_RATE_{str(update_rate)}')
X = merged_dataframe[X_cols]
y = merged_dataframe['HOME_TEAM_WINS']
X_test, X_train = X[:int(len(merged_dataframe)*.1)], X[int(len(merged_dataframe)*.1):]
y_test, y_train = y[:int(len(merged_dataframe)*.1)], y[int(len(merged_dataframe)*.1):]
X_train
```

```python
max_depths = [None, 20, 10, 5]
min_samples_leaf_list = [1, 5, 10]
best_score = 0
for max_depth in max_depths:
    for min_samples_leaf in min_samples_leaf_list:
        clf=RandomForestClassifier(max_depth=max_depth, min_samples_leaf=min_samples_leaf)
        clf.fit(X_train, y_train)
        print(f"Scores for tree with max_depth of {max_depth} and min_samples_leaf of {min_samples_leaf}")
        print (clf.score(X_train, y_train))
        print(clf.score(X_test, y_test))
        if clf.score(X_test, y_test) > best_score:
            best_params = [max_depth, min_samples_leaf]
            best_score = clf.score(X_test, y_test)
print(f"The best parameters found were a max_depth of {best_params[0]} and min_samples_leaf of {best_params[1]}")
print(best_score)
# achieved 63.15% accuracy with max_depth of 10 & min_samples_leaf of 1
```
