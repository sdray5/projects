```python
import pandas as pd
lol_games = pd.read_csv("high_diamond_ranked_10min.csv")
lol_games
```

```python
import numpy as np
lol_games.isnull().any()
# no missing data
```

```python
lol_games = lol_games.drop('gameId', axis=1)
lol_games
```

```python
from sklearn.preprocessing import StandardScaler
lol_games_standardized = pd.DataFrame(StandardScaler().fit_transform(lol_games), columns = lol_games.columns)
lol_games_standardized
```

```python
np.sum(abs(lol_games_standardized) > 3, axis=0)/len(lol_games_standardized)*100
```

```python
np.sum(abs(lol_games_standardized) > 5, axis=0)/len(lol_games_standardized)*100
```
